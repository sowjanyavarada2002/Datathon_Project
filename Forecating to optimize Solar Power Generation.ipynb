# Data Modeling & Evaluation

# Predicting Clearsky GHI

```python
final_train_ghi = final_train.drop('Clearsky DHI', axis = 1)
final_train_ghi = final_train_ghi.drop('Solar Zenith Angle', axis = 1)
```

```python

final_train_ghi = final_train_ghi.drop('Clearsky DNI', axis = 1)
y_train_GHI = final_train_ghi.pop('Clearsky GHI')
X_train_GHI = final_train_ghi
```

```python
y_train_GHI
```

```python
X_train_GHI
```

```python
#import RFE and feature Elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train_GHI, y_train_GHI)
```

```python
rfe = RFE(lm,20)
rfe = rfe.fit(X_train_GHI, y_train_GHI)
```

```python
list(zip(X_train_GHI.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train_GHI.columns[rfe.support_]
col
```

```python
X_train_GHI.columns[~rfe.support_]
```

## Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train_GHI[col]
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
lm =sm.OLS(y_train_GHI, X_train_rfe).fit()
#Let's see the summary of our linear model
print(lm.summary())

```

```python
X_train_new = X_train_rfe.drop(['Precipitable Water'], axis=1)
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train_new)
lm=sm.OLS(y_train_GHI, X_train_lm).fit()
print(lm.summary())
```

```python
# All the p values are less than 0.05
```
